```python
import os
from plot_classes import *

%load_ext line_profiler
```

```python
mydict = {
    'font.size': 10
}

lbls = ['SM','SQR_ctp', 'SQR_ctpI','SQR_ctW','SQR_ctWI']#,'SQR_cpt', 'SQR_ctG','SQR_ctGI','SQR_ctZ','SQR_ctZI','SQR_cpQM','SQR_cpt']

smples = ['TTH_{}_Inclusive_Nominal'.format(x) for x in lbls]
legends = {smples[x]:lbls[x] for x in range(len(lbls))}

plot = Hist1D(
    observable='top_pt',
    samples = smples,
    data=['TTH_SM_Inclusive_Nominal', 'TTH_SQR_ctW_Inclusive_Nominal'],
    stack=False,
    errors='hist',
    directory = os.getcwd() + '/../../PythiumTest2/output/',
    labels=legends,
    logy=True,
    ylim=(1e2,1e5)
)

plot.figure_options(figsize=(5,5), mastertitle="Master Title") 
plot.plot_options(rcp_kw=mydict, gridstring='y:',markersize=3)
plot.color_options(colormap='plasma', reverse=True)
plot.set_axislabels(12, ymain="Events", xmain=plot.labels[0])
plot.fontsize_options(ticks=10)
plot.create()
plot.saveimage("Hist1D_demo_plasma")
```

```python
plot = RatioPlot(
    observable = 'top_pt',
    samples=smples, 
    data=['TTH_SM_Inclusive_Nominal','TTH_SQR_ctW_Inclusive_Nominal',],
    reference='TTH_SM_Inclusive_Nominal',
    stack=False,
    errors='hist',
    directory= os.getcwd() + '/../../PythiumTest2/output/',
    labels=legends,
)
plot.figure_options(spacing=0.02, figsize=(5,6))
plot.ratio_options([0.5,1.5], 0.25)
plot.plot_options(shape='hollow',markersize=3,logx=True,xlim=(1,1000), legendloc='center left')
plot.set_axislabels(ymain="Events", xbot=plot.names[0])
plot.fontsize_options(labels=13, ticks=10)
plot.create()
```
